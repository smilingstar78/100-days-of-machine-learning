```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Flatten
from keras.applications.vgg16 import VGG16
import os
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
folder_path = '/content/utkface_aligned_cropped/UTKFace'
```

```python
age=[]
gender=[]
img_path=[]
for file in os.listdir(folder_path):
  age.append(int(file.split('_')[0]))
  gender.append(int(file.split('_')[1]))
  img_path.append(file)
```

```python
len(age)
```

```python
df = pd.DataFrame({'age':age,'gender':gender,'img':img_path})
```

```python
df.shape
```

```python
df.head()
```

```python
train_df = df.sample(frac=1,random_state=0).iloc[:20000]
test_df = df.sample(frac=1,random_state=0).iloc[20000:]
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=30,
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=folder_path,
    x_col='img',
    y_col=['age','gender'],
    target_size=(224,224),
    class_mode='raw'
)

test_generator = test_datagen.flow_from_dataframe(
    dataframe=test_df,
    directory=folder_path,
    x_col='img',
    y_col=['age','gender'],
    target_size=(224,224),
    class_mode='raw'
)
```

```python
def multi_output_generator(generator):
    for x, y in generator:
        # y[:,0] = age, y[:,1] = gender
        yield x, {"age": y[:,0], "gender": y[:,1]}
train_gen = multi_output_generator(train_generator)
test_gen = multi_output_generator(test_generator)
```

```python
conv_base = VGG16(
    weights='imagenet',
    include_top = False,
    input_shape=(150,150,3)
)
```

```python
conv_base.trainable = True

set_trainable = False

for layer in conv_base.layers:
  if layer.name == 'block5_conv1':
    set_trainable = True
  if set_trainable:
    layer.trainable = True
  else:
    layer.trainable = False

for layer in conv_base.layers:
  print(layer.name,layer.trainable)
```

```python
conv_base.summary()
```

```python
from keras.models import Model
```

```python
x = keras.layers.Input(shape=(224,224,3))

flat = keras.layers.Flatten()(x)
hidden1 = Dense(128, activation='relu')(flat)
hidden2 = Dense(64, activation='relu')(hidden1)

age_output = keras.layers.Dense(1, activation='linear', name="age")(hidden2)
gender_output = keras.layers.Dense(1, activation='sigmoid', name="gender")(hidden2)
```

```python
model = Model(inputs = x ,outputs = [age_output,gender_output])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model,show_shapes=True)
```

```python
model.compile(optimizer='adam', loss={'age': 'mae', 'gender': 'binary_crossentropy'}, metrics={'age': 'mae', 'gender': 'accuracy'},loss_weights={'age':1,'gender':99})
```

```python
model.fit(train_gen,
          validation_data=test_gen,
          batch_size = 32,
          epochs=10,
          steps_per_epoch=len(train_generator),
          validation_steps=len(test_generator))
```
